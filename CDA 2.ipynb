# Calculating the uniform cdf using scipy.stats 
by importing uniform and specifying the values inside uniform.cdf ie., x,location and scale after that printed cdf upto range of 1000

```python
# Cdf 
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
min_value = 20
max_value = 40
x=np.linspace(20,40,1000)
cdf_calc = uniform.cdf(x,loc=min_value,scale = (40-20))
print(cdf_calc)
```

# Plotting the cdf uniform distribution 
in betweem the range of 20 to 40 and plotted the graph using matplotlib and labelled x and y axis as below 

```python
p = uniform.cdf(x,20,20)
plt.plot(x,p)
plt.title ("cdf uniform distribution")
plt.xlabel ("temperature")
plt.ylabel ("cdf")
```

# Binomial Distribution
varyfying the concept of binomial distribution in scipy.stats by importing binom then specifying the values inside the function binom.pmf printed the same using the bino variable 

```python
#Binomial Distribution

import numpy as np
from scipy.stats import binom
bino = binom.pmf (3,5,0.6)
print(bino)
```

# Geometric Distribution 
similarly to the above the values specified inside the function geom.pmf for geometric calc and printed the same using geo variable

```python
import numpy as np
from scipy.stats import geom
geo= geom.pmf(4,.6)
print(geo)
```

# Uniform CDF
uniform cdf calculated by substracting the cdf_1 from cdf_1 as code below

```python
# uniform cdf
import numpy as np
from scipy.stats import uniform
cdf_1= uniform.cdf(28, loc=25, scale=10)
cdf_2= uniform.cdf(32, loc=25, scale=10)
print(cdf_2)
print(cdf_1)
print(cdf_2 - cdf_1)
```
