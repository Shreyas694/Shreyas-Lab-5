#### C.Probability of exactly 2 rainy days in 5 day period 

```python
# Import necessary libraries 
import numpy as np 
from scipy.stats import binom 

# Import the values 
n = 5 
p = 0.3 
k = 2 
prb = binom.pmf(k,n,p)  
print(f"Probability of exactly 2 rainy days in a 5 day-period = { prb:.4f}") 
```

### Q2.Probability Distribution 
#### A. Generating a Poisson distribution  (given size = 1000 , average (mu) = 12)

```python
# Import neccessart libraries 
import numpy as np 
import random
from scipy.stats import poisson
mu = 12
k = 1000
random_sample =  np.random.randint(12,24,1000)
print(f"Poisson distributed random sample = {random_sample}")
```

#### B. Plotting the histogram

```python
import matplotlib.pyplot as plt 
plt.hist(random_sample,bins = 20, density=True,edgecolor='black',color='skyblue',alpha=0.7)
plt.title("Histogram plot")
plt.xlabel("range")
plt.ylabel('hight')
plt.grid(True)
plt.show()
```

#### C.  Calculating Mean and Varience 

```python
mu = 12
k = 1000
prb = poisson.mean(k,mu)
print(f"Poisson distributed mean of random sample = {prb}")

prb1 = poisson.var(k,mu)
print(f"Pisson distributed Variance of random sample = {prb1}")
```

### D. Interpretation 
The Poisson model shows us the exact rainfall events over a large number of opportunities

#### D. Given lat = 15-25N, lon = 75-85E for the year 2010 & plotting the JJAS anamoly Time Series

```python
# Import the neccessary library
import numpy as np 
import matplotlib.pyplot as plt 
import xarray as xr 
import glob 
path = 'D:\TP_data\*.nc'
ds = xr.open_mfdataset(path)
print(ds)

rain = ds['rf']

lat1 = 15
lon1 = 75
lat2 = 25
lon2 = 85

print(rain)

rain_jjas_2010 = rain.sel(time=slice('2010-06-01', '2010-09-30'))

rr1 = rain.sel(lat=lat1,lon=lon1,method = 'nearest')
rr2 = rain.sel(lat=lat2,lon=lon2,method = 'nearest')
print(rr1,rr2)

rr1.plot(color = 'blue',label ='Location_1')
rr2.plot(color = 'red',label = 'Location_2')
plt.title("Rainfall Data (June - September 2010)") 
plt.xlabel("Date")
plt.ylabel("Rainfall (mm/day)")
plt.legend()
plt.grid(True)
plt.show()
```

### Q4.Spatial 
#### 4. Positive and Negative values 

```python
# Removing zero rainfall 
rain_pos = rr1.values[rr1.values>0]
print(rain_pos)

# For positive rainfall 
rain_log = np.log(rain_pos)
print(rain_log)

fig, ax = plt.subplots(1, 2, figsize=(12,4))

# Histogram of positive rainfall
edges_pos = np.arange(0, np.nanmax(rain_pos) + 2, 2)
ax[0].hist(rain_pos, bins=edges_pos, edgecolor='black', color='skyblue')
ax[0].set_title('Positive Rainfall (2-mm bins)')
ax[0].set_xlabel('Rainfall (mm)')
ax[0].set_ylabel('Frequency')
ax[0].set_xticks(edges_pos[::4])
ax[0].set_xticklabels(edges_pos[::4], rotation=45)

ax[0].set_ylabel('Frequency')
ax[0].set_xticks(edges_pos[::4])
ax[0].set_xticklabels(edges_pos[::4], rotation=45)

# Histogram of log rainfall
log_edges = np.linspace(np.nanmin(rain_log), np.nanmax(rain_log), 30)
ax[1].hist(rain_log, bins=log_edges, edgecolor='black', color='salmon')
ax[1].set_title('Log(Rainfall > 0 mm)')
ax[1].set_xlabel('log(mm)')
ax[1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```
